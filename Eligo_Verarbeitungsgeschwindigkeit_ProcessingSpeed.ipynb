```python
#basic_untimed version

import random
import time

# Define the symbols
symbols = ['●', '▲', '♣', '♦', '▼', '✚', '♥', '⬟', '⬢']

def assign_values():
    random.shuffle(symbols)
    symbol_values = {symbols[i]: i + 1 for i in range(9)}
    return symbol_values

def print_symbols(symbol_values):
    print("Symbols and their values:")
    for i in range(1, 10):
        symbol = list(symbol_values.keys())[list(symbol_values.values()).index(i)]
        print(f"{symbol}: {i}", end=" ")
    print("\n")

def generate_symbol(symbol_values):
    return random.choice(list(symbol_values.keys()))

def generate_question(symbol_values):
    symbol1 = generate_symbol(symbol_values)
    symbol2 = generate_symbol(symbol_values)
    while symbol1 == symbol2:
        symbol2 = generate_symbol(symbol_values)
    return symbol1, symbol2

def main():
    while True:
        symbol_values = assign_values()
        print_symbols(symbol_values)

        symbol1, symbol2 = generate_question(symbol_values)
        value1 = symbol_values[symbol1]
        value2 = symbol_values[symbol2]

        task = random.choice(["larger", "smaller"])
        print(f"Which has the {task} value?")
        print(f"1. {symbol1}   or   2. {symbol2}")
        user_input = input("Your answer (1 or 2): ")

        if task == "larger":
            correct_answer = "1" if value1 > value2 else "2"
        else:
            correct_answer = "1" if value1 < value2 else "2"

        if user_input == correct_answer:
            print("Correct!\n")
        else:
            print(f"Wrong. The correct answer is {correct_answer}\n")

        time.sleep(0.1)

if __name__ == "__main__":
    main()
```

```python
###Timed_Verarbeitungsgeschwindigkeit -- Let's Crack Eligo2

'''
Total 20 Questions - Random Generation
Every 10 Second, a new question
Correct Answer = +1
Wrong Answer = -1
Unanswered = -1
'''

import random
import time
import threading
import sys

# Define the symbols
symbols = ['●', '▲', '♣', '♦', '▼', '✚', '♥', '⬟', '⬢']

def assign_values():
    random.shuffle(symbols)
    symbol_values = {symbols[i]: i + 1 for i in range(9)}
    return symbol_values

def print_symbols(symbol_values):
    print("Symbols and their values:")
    for i in range(1, 10):
        symbol = list(symbol_values.keys())[list(symbol_values.values()).index(i)]
        print(f"{symbol}: {i}", end=" ")
    print("\n")

def generate_symbol(symbol_values):
    return random.choice(list(symbol_values.keys()))

def generate_question(symbol_values):
    symbol1 = generate_symbol(symbol_values)
    symbol2 = generate_symbol(symbol_values)
    while symbol1 == symbol2:
        symbol2 = generate_symbol(symbol_values)
    return symbol1, symbol2

def ask_question(symbol_values, score, question_number):
    print(f"\nQuestion {question_number}:")
    print_symbols(symbol_values)

    symbol1, symbol2 = generate_question(symbol_values)
    value1 = symbol_values[symbol1]
    value2 = symbol_values[symbol2]

    task = random.choice(["larger", "smaller"])
    print(f"Which has the {task} value?")
    print(f"1. {symbol1}   or   2. {symbol2}")

    # To capture input with timeout
    user_input = [None]
    def get_input():
        user_input[0] = input("Your answer (1 or 2): ")

    input_thread = threading.Thread(target=get_input)
    input_thread.start()
    input_thread.join(timeout=10)

    if user_input[0] is None:
        print("\nTime's up!")
        sys.stdout.write("\033[F")  # Cursor up one line
        sys.stdout.write("\033[K")  # Clear line
        sys.stdout.write("\033[F")  # Cursor up one line
        sys.stdout.write("\033[K")  # Clear line
        score -= 1
    else:
        correct_answer = "1" if (task == "larger" and value1 > value2) or (task == "smaller" and value1 < value2) else "2"

        if user_input[0] == correct_answer:
            print("Correct!")
            score += 1
        else:
            print(f"Wrong. The correct answer is {correct_answer}")
            score -= 1

    return score

def main():
    score = 0
    total_questions = 20

    for question_number in range(1, total_questions + 1):
        symbol_values = assign_values()

        score = ask_question(symbol_values, score, question_number)
        time.sleep(1)  # Short delay before the next question

    print(f"\nFinal Score: {score} out of {total_questions}")

if __name__ == "__main__":
    main()
```
